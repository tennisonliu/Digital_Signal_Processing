## Define the play_audio function

```python
def play_audio( Q, p, fs , dev=None):
    # play_audio plays audio with sampling rate = fs
    # Q - A queue object from which to play
    # p   - pyAudio object
    # fs  - sampling rate
    # dev - device number
    
    # Example:
    # fs = 44100
    # p = pyaudio.PyAudio() #instantiate PyAudio
    # Q = Queue.queue()
    # Q.put(data)
    # Q.put("EOT") # when function gets EOT it will quit
    # play_audio( Q, p, fs,1 ) # play audio
    # p.terminate() # terminate pyAudio
    
    # open output stream
    ostream = p.open(format=pyaudio.paFloat32, channels=1, rate=int(fs),output=True,output_device_index=dev)
    # play audio
    while (1):
        data = Q.get()
        if data is "EOT" :
            break
        try:
            ostream.write( data.astype(np.float32).tostring() )
        except:
            break
```

## Sample Sinusoidal Signals
### Case 1: 5 kHz sampled with 72 kHz - this will be our reference

```python
# sampled at 5 Hz
print(np.r_[0:5]*(5/5))
# sampled at 10 Hz
print(np.r_[0:5]*(10/5))
```

```python
from __future__ import division
fs = [72000, 24000, 12000, 8000, 6000]

# Case 1: 5 kHz sampled with 72 kHz
# Create signal - 2 second signal sampled at fs
t1 = np.r_[0:2:(1/fs[0])]
sig1 = np.sin(2*np.pi*5000*t1)

# Perform FFT and create frequency axis
# 1024 point DFT
sig1fft = np.fft.fft(sig1, 1024 )
freq1 = np.r_[0:1024]*(fs[0]/1024)

# Plot the signal 
plt.figure()
width, height = plt.figaspect(0.2)
fig = plt.figure(figsize=(width,height))
# plot 0.01 second, convert to 10 ms scale
plt.plot(t1[0:720]*1000,sig1[0:720])
plt.legend(("fs = 72 kHz",""))
plt.xlabel("time (ms)")
plt.ylabel("Amplitude")

# Plot the FFT of the signal
plt.figure()
width, height = plt.figaspect(0.2)
fig = plt.figure(figsize=(width,height))
# plot the first 512 frequency bins, scaled to kHz range, normalise FFT
plt.plot(freq1[0:513]/1000,abs(sig1fft[0:513])/np.max(abs(sig1fft[0:513])))
plt.xlabel("frequency (kHz)")
plt.ylabel("Normalized Amplitude")
plt.legend(("fs = 72 kHz",""))
plt.axis([0,10,0,1])
plt.xticks(np.arange(0,10+1,1))
```

## Play the reference signal

```python
Qout = queue.Queue()
Qout.put( sig1 );
Qout.put( "EOT" );
p = pyaudio.PyAudio()
play_audio(Qout,p,fs[0])
p.terminate()
```

## Comment on the Previous Exercises

Since we are using 1024 FFT length, that means the 1024 point FFT will always contain 512 output frequency bins. The frequency resolution of each frequency bin is given by (sampling rate/FFT size). We observe that as the sample rate we use becomes smaller and smaller, the frequency resolution of each frequency bin decreases as well (i.e. from 72000/1024 = 70.3 Hz to 6000/1024 = 5.86Hz). As the frequency resolution improves, the FFT curves become smoother and more accurate. 

However, in the time domain, as less samples are taken on the 5kHz signal, more information is lost in the time domain. This also limits the available bandwidth of the signal, as the Nyquist Sampling Theorem states that the sampling frequency should be at least twice the largest frequency component in the signal. In our case, the sampling frequency should be at least 10 kHz. In the last two examples of using sampling frequencies of 8kHz and 6kHz, we see that the frequency spectrum is incorrect, suggesting that aliasing has occured due to insufficient sampling rate.

This suggests that a tradeoff will have to be made when trying to decide between frequency resoluiton and time resolution.
